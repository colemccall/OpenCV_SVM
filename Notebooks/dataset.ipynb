```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plot
import pickle
print("Imports complete")
```

```python
train_dir = os.path.join(os.path.normpath(os.getcwd() + os.sep + os.pardir), 'Data\\Training')
print("Training Data Directories are located at " + train_dir)
#'G:\\Shared drives\\FireMAP\\Summer_2022_Imagery\\EastTroublesome\\Worldview3\\2021_06_04_Worldview_3\\21JUN04180514-M3DS_R1C1-014175810010_01_P001.tif'
```

```python
classes = ['Tree', 'Not_Tree']

data = []

# Have one directory for each label
for label in classes:
    # Get the path and label of each training data directory
    path = os.path.join(train_dir, label)
    current_label = classes.index(label)
    print("Files included within " + path + " belong to class " + label)
    count = 0

    # For each image in the directory
    for image in os.listdir(path):
        # Load the path of each image
        image_path = os.path.join(path, image)
        print('\t' + image_path) ; count+=1

        # Read the image using OpenCV, then resize it
        test_image=cv2.imread(image_path,0)
        try:
            test_image=cv2.resize(test_image, (50,50))

            # 2D array to 1D array transformation
            current_image = np.array(test_image).flatten()

            # Create a list of objects that contains the image and class label
            data.append([current_image, current_label])
            #cv2.imshow('image', test_image)
        except Exception as e:
            pass
    
    # Verify that the images were read in
    print(str(count) + " " + label + " images were read in\n\n")



# Save the data using pickle
pick_in = open('..\\Data\\Pickle\\data1.pickle', 'wb')
pickle.dump(data, pick_in)
pick_in.close()
```

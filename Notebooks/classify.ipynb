Imports

```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plot
import pickle
import random
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC as SVM
print("Imports complete")
```

Using pickle, load in a trained SVM that was created in the `Train` Notebook

```python
pickle_model_path = os.path.join(os.path.normpath(os.getcwd() + os.sep + os.pardir), 'Data\\Pickle\\model.sav')
print("Model located at " + pickle_model_path)
pick_in = open(pickle_model_path,'rb')
model = pickle.load(pickle_model_path)
pick_in.close()

pickle_dataset = os.path.join(os.path.normpath(os.getcwd() + os.sep + os.pardir), 'Data\\Pickle\\data1.pickle')
print("Dataset located at " + pickle_dataset)
pick_in = open(pickle_dataset,'rb')
data = pickle.load(pickle_dataset)
pick_in.close()
```

Split Dataset into a Training Set and a Testing Set

```python
random.shuffle(data)

features = []
labels = []

for feature, label in data:
    features.append(feature)
    labels.append(label)

# Use 80% of the dataset for training and 20% for testing
test_set = 0.1
train_set = 0.9
xtrain, xtest, ytrain, ytest = train_test_split(features, labels, test_size=test_set, train_size=train_set)
```

```python
prediction = model.predict(xtest)

accuracy = model.score(xtest, ytest)

classes = ['Tree, Not_Tree']
print('Accuracy: ', accuracy)
print('Prediction is :', classes[prediction[0]])
```
